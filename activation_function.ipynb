## Activation Functions - All You Need To Know
For explanation to the code, refer to https://mldoodles.com/activation-functions-all-you-need-to-know/

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp = [-2, -1, 0, 1, 2]

# hidden layer weights and biases
w1, b1 = 2, 1  # This is equivalent to the equation y = 2x+1
w2, b2 = 3, -2 # This is equivalent to the equation y = 3x-2
```

```python
# This represents first layer
def n1(x):
    x = np.array(x)
    w1, b1 = 2, 1
    y = []
    for i in x:
        y.append(w1*i + b1)
    return y

# This represents second layer
def n2(x):
    x = np.array(x)
    w2, b2 = 3, -2
    y = []
    for i in x:
        y.append(w2*i + b2)
    return y
```

```python
def sig(x):
    x = np.array(x)
    sigmoid = (1 / (1+ np.exp(-x)))        
    return list(sigmoid)

def tanh(x):
    x = np.array(x)
    tanh = (np.exp(x)-np.exp(-x)) / (np.exp(x)+np.exp(-x))
    return list(tanh)

# relu => y = max(0, x)
def relu(x):
    y = []
    for i in x:
        y.append(max(0, i))
    return y
```

```python
# Without Activation Layer
n1_out = n1(inp)
n2_out = n2(n1_out)
```

```python
plt.figure(figsize=(14,5))

plt.subplot(121)
plt.axis([-2, 2, -15, 15])
plt.plot(inp, n1_out)
plt.xlabel("input")
plt.ylabel("neuron1 output")
plt.title("layer 1")

plt.subplot(122)
plt.axis([-2, 2, -15, 15])
plt.plot(inp, n2_out)
plt.xlabel("input")
plt.ylabel("neuron2 output")
plt.title("layer 2")

plt.suptitle("Without Activation Function")
plt.show()
```

```python
# With Activation - Sigmoid
n1_out = n1(inp)
sig_n1_out = sig(n1_out)
n2_out = n2(sig_n1_out)
sig_n2_out = sig(n2_out)
```

```python
plt.figure(figsize=(14,5))

plt.subplot(121)
plt.axis([-2, 2, 0, 1.2])
plt.plot(inp, sig_n1_out)
plt.xlabel("input")
plt.ylabel("layer1 sigmoid output")
plt.title("layer 1")

plt.subplot(122)
plt.axis([-2, 2, 0, 1.2])
plt.plot(inp, sig_n2_out)
plt.xlabel("input")
plt.ylabel("layer2 sigmoid output")
plt.title("layer 2")

plt.suptitle("Activation function - sigmoid")
plt.show()
```

```python
# With Activation - tanh
n1_out = n1(inp)
tanh_n1_out = tanh(n1_out)
n2_out = n2(tanh_n1_out)
tanh_n2_out = tanh(n2_out)
```

```python
plt.figure(figsize=(14,5))

plt.subplot(121)
plt.axis([-2, 2, -1.25, 1.25])
plt.plot(inp, tanh_n1_out)
plt.xlabel("input")
plt.ylabel("layer1 tanh output")
plt.title("layer 1")

plt.subplot(122)
plt.axis([-2, 2, -1.25, 1.25])
plt.plot(inp, tanh_n2_out)
plt.xlabel("input")
plt.ylabel("layer2 tanh output")
plt.title("layer 2")

plt.suptitle("Activation function - tanh")
plt.show()
```

```python
# With Activation - ReLU
n1_out = n1(inp)
relu_n1_out = relu(n1_out)
n2_out = n2(relu_n1_out)
relu_n2_out = relu(n2_out)
```

```python
plt.figure(figsize=(14,5))

plt.subplot(121)
plt.axis([-2, 2, -2, 14])
plt.plot(inp, relu_n1_out)
plt.xlabel("input")
plt.ylabel("layer1 relu output")
plt.title("layer 1")

plt.subplot(122)
plt.axis([-2, 2, -2, 14])
plt.plot(inp, relu_n2_out)
plt.xlabel("input")
plt.ylabel("layer2 relu output")
plt.title("layer 2")

plt.suptitle("Activation function - relu")
plt.show()
```
